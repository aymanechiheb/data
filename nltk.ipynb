```python
def rule_based_pos_tagging(sentence):
    words = sentence.split()
    pos_tags = []
    for word in words:
        if word.endswith('ing'):
            pos_tags.append((word, 'VBG'))  # Verb, gerund/present participle
        elif word.endswith('ly'):
            pos_tags.append((word, 'RB'))   # Adverb
        else:
            pos_tags.append((word, 'NN'))   # Noun
    return pos_tags



```

```python
# Test the rule-based POS tagging on multiple sentences
sentences = ["He is running quickly.", "She sings beautifully."]
for sentence in sentences:
    print(rule_based_pos_tagging(sentence))
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
def nltk_pos_tagging(sentence):
    words = nltk.word_tokenize(sentence)
    pos_tags = nltk.pos_tag(words)
    return pos_tags
 
# Compare results with rule-based approach
for sentence in sentences:
    print("Rule-based POS tagging:", rule_based_pos_tagging(sentence))
    print("NLTK POS tagging:", nltk_pos_tagging(sentence))
    print()
```

```python
def improved_rule_based_pos_tagging(sentence):
    words = sentence.split()
    pos_tags = []
    for word in words:
        if word.endswith('ing'):
            pos_tags.append((word, 'VBG'))  # Verb, gerund/present participle
        elif word.endswith('ly'):
            pos_tags.append((word, 'RB'))   # Adverb
        elif word.endswith('ed'):
            pos_tags.append((word, 'VBD'))  # Verb, past tense
        elif word.endswith('s'):
            pos_tags.append((word, 'NNS'))  # Noun, plural
        else:
            pos_tags.append((word, 'NN'))   # Noun (default)
    return pos_tags

# Test the improved rule-based POS tagging
for sentence in sentences:
    print("Improved rule-based POS tagging:", improved_rule_based_pos_tagging(sentence))
```
